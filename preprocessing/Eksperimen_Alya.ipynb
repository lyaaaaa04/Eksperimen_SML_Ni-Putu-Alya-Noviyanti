# **1. Perkenalan Dataset**


Dalam implementasi ini, saya menggunakan dataset Students Performance in Exams yang diperoleh dari **Kaggle**. Dataset ini berisi informasi mengenai prestasi siswa dalam berbagai ujian mata pelajaran, serta data mengenai latar belakang pribadi dan pendidikan orang tua mereka.

Dataset ini memiliki 8 kolom utama, antara lain:


1.   gender: Menunjukkan jenis kelamin siswa (male/female)
2.   race/ethnicity: Klasifikasi siswa berdasarkan ras atau etnis (misal: group A, group B, dsb.)
3.   parental level of education: Tingkat pendidikan yang telah dicapai oleh orang tua siswa
4.   lunch: Jenis layanan makan siang yang diterima siswa (standard/reduced)
5.   test preparation course: Status keikutsertaan siswa dalam kursus persiapan ujian (completed/none)
6.   math score: Skor ujian matematika (0–100)
7.   reading score: Skor ujian membaca (0–100)
8.   writing score: Skor ujian menulis (0–100)

Dataset ini berisi sekitar 1000 baris, di mana setiap baris mewakili seorang siswa beserta hasil ujian mereka. Data ini berguna untuk menganalisis faktor-faktor yang memengaruhi prestasi akademik siswa dan dapat digunakan untuk membangun model prediktif guna memperkirakan hasil ujian berdasarkan data demografis dan kebiasaan belajar.

```python
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
```

```python
def score_to_grade(score):
    if score >= 85:
        return 'A'
    elif score >= 75:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'E'

df['performance_level'] = df['average_score'].apply(score_to_grade)

df.head()
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
for outlier in numeric_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[outlier])
    plt.title(f'Box Plot of {outlier}')
    plt.show()
```

```python
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
df_encoded = pd.get_dummies(df, columns=categorical_cols)
```

```python
X = df_encoded.drop(['math score', 'reading score', 'writing score', 'average_score', 'performance_level'], axis=1)
y = df_encoded['performance_level']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
